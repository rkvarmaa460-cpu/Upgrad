```python
import pandas as pd


sampled_data_jan = pd.DataFrame()

df1['tpep_pickup_datetime'] = pd.to_datetime(df1['tpep_pickup_datetime'])
df1['date'] = df1['tpep_pickup_datetime'].dt.date
df1['hour'] = df1['tpep_pickup_datetime'].dt.hour

# Loop through dates
for date in df1['date'].unique():
    date_data = df1[df1['date'] == date]

    # Loop through all 24 hours
    for hour in range(24):
        hour_data = date_data[date_data['hour'] == hour]

        if len(hour_data) > 0:
            sample = hour_data.sample(frac=0.05, random_state=42)
            sampled_data_jan = pd.concat([sampled_data_jan, sample])
```

```python
def time_of_day(hour):
  if hour>=5 and hour<=12:
    return 'Morning'
  elif hour> 12 and hour<=17:
    return 'Afternoon'
  elif hour>17 and hour<21:
    return 'Midnight'
  else:
    return 'Night'

```
